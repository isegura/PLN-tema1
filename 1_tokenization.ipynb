# Ejercicios sobre Tokenización


Implementa un código (en Python) que permita tokenizar el siguiente texto:

*Hola, ¿cómo estás? Estoy bien :) ¿Y tú? Escríbeme a isegura@inf.uc3m.es o visita https://hulat.inf.uc3m.es/nosotros/miembros/isegura.
*

La salida esperada debe ser algo como:
["Hola,", "¿cómo", "estás?", "Estoy", "bien", ":)", "¿Y", "tú?", "Escríbeme", "a", "isegura@inf.uc3m.es", "o", "visita",  "https://hulat.inf.uc3m.es/nosotros/miembros/isegura" , “.”]


```python
text="Hola, ¿cómo estás? Estoy bien :) ¿Y tú? Escríbeme a isegura@inf.uc3m.es o visita https://hulat.inf.uc3m.es/nosotros/miembros/isegura."
```

```python
tokens = text.split()
print(tokens)
```

**¿Qué problemas detectas??**

- Algunas palabras no se separan correctamente de los signos de puntuación (por ejemplo, "estás?").
- La URL incluye el punto final,


## Tokenización con nltk


```python
import nltk
from nltk.tokenize import word_tokenize

# Necesitamos descargar algunos modulos de nltk
nltk.download('punkt')
nltk.download('punkt_tab')


tokens = word_tokenize(text, language="spanish")

print(tokens)
```
